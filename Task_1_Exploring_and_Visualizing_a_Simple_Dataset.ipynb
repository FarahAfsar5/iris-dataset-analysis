```python
# Import necessary libraries
import pandas as pd               # For data manipulation and analysis
import seaborn as sns             # For statistical data visualization
import matplotlib.pyplot as plt   # For general plotting

# Load the Iris dataset using seaborn (you can also load it from a CSV file if needed)
iris = sns.load_dataset('iris')

# -------------------- Data Inspection --------------------

# 1. Display the number of rows and columns in the dataset
print("Dataset Shape (rows, columns):", iris.shape)

# 2. Display the column names
print("\nColumn Names:")
print(iris.columns)

# 3. Display the first 5 rows of the dataset
print("\nFirst 5 Rows of the Dataset:")
print(iris.head())

# -------------------- Data Visualization --------------------

# Set the style of seaborn plots
sns.set(style="whitegrid")

# 1. Scatter Plot: Analyze relationships between two variables
# Example: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()

# 2. Histogram: Examine the distribution of Petal Length
plt.figure(figsize=(8, 6))
sns.histplot(data=iris, x='petal_length', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# 3. Box Plot: Detect outliers and understand value spread for each species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```
